Note la diferencia entre `==` (igualdad entre vairables) and `=` (asignación de variable)

```python
z = 3

print("Z = {}".format(z))
print("Es igual (==) a 2 ? {}  ; Es igual a 3 ? {}".format(z == 2, z == 3))
print("\nMayor (>) que 2 ? {}".format(z>2))
print("\nMayor (>) que 0.5 pero menor (<) que 4 ? {}".format(0.5 < z <= 4))
print("\nEs diferente (!=) a 5 ? {}".format(z!=5))
```

### Herramientas de Control de Flujo

```python
# IF : Si [argumneto verdad]: ejecutar codigo ... si no: ejecture otro codigo
colombia = ["Cali", "Medellin"]
escocia = ["Edimburgo", "Glasgow"]
test = "Cali"
# test = "Edimburgo"
# test = "Miami"

if test in colombia:
    print("{} si esta en Colombia".format(test))
    
elif test in escocia:
    print("{} si esta en Escocia".format(test))
    
else:
    print("{} no esta en estos paises ".format(test))
```

```python
# FOR : util para recorrer una lista o generar ciclos

colombia = ["Cali", "Chipichape", "Yumbo"]

for ciudad in colombia:
    print(ciudad)
```

```python
escocia = ["Edimburgo", "Glasgow", "Aberdeen"]
kill = False
# kill = True

for i in range(len(escocia)):
    print(escocia[i])
    
    if kill: 
        print("BREAK !")
        break
```

### Numpy 

Numpy permite crear y manipular [vectores multidimensionales](http://docs.scipy.org/doc/numpy/reference/arrays.ndarray.html)

```python
# Importar la libreria numpy como una abreviación "np". Podemos usar las funciones de Numpy
# llamando np.<nombre_de_funcion> 
import numpy as np

array1 = np.array([
    [1, 2, 3],
    [4, 5, 6]]
)

print(array1.shape)  # np.shape indica las dimensiones del vector multidimensional
array1
```

Los elementos del vector se pueden obtener via la [sintaxis de indices](http://docs.scipy.org/doc/numpy/reference/arrays.indexing.html#arrays-indexing) (NOTA: La primera posicion es la posicion 0)

```python
print("Vector completo : \n {} \n".format(array1))
print("Elemento (2,3) : ", array1[1,2])
print("Primera fila y todas las columnas : ", array1[0,:])
print("Segunda columna y todas las filas : ", array1[:,1])
print("Todos los elementos de la segunda fila, comenzando en la 2da columna : ", array1[1, 1:])
```

Existen [multiples funciones](http://docs.scipy.org/doc/numpy/reference/) para manipular y obtener información importante de los vectores.

```python
print(array1.mean(axis=1)) # Media a lo largo de las filas
print(array1.mean(axis=0)) # Media a lo largo de las Columnas
print(array1.ravel()) # "Aplanar" el vector a lo largo de una sola dimension
print(array1.reshape((3,2))) # Cambiar las dimensiones del arreglo
```

Podemos usar los operadores de comparacion para "filtrar" que elementos del vector cumplen con determinada condicion ... y podemos usar la **mascara** para indexiar el vector :

```python
mask = (array1 > 2) & (array1 < 5)
print(array1)
print(mask)
print(array1[mask])
```

El poder de Numpy se ve en las **operaciones matriciales**, con las cuales podemos multiplicar, sumar, calcular operaciones tensoriales etc.

```python
x = np.array([1,2,3])
y = np.array([2,3,4])
z = np.dot(x,y)
print(x)
print(y)
print(z)
```

```python
# Inicializar vectores
D = 3
H = 2
model = {}
model['W1'] = np.random.randn(H,D) # Dimensiones: H x D
model['W2'] = np.random.randn(H) # Dimensions: H x 1
print(model['W1'])
print(model['W2'])

# Entrada
x = np.array([1,2,3]) # Dimensiones: 3 x 1 == D X 1

# Feedforward
act_1 = np.dot(model['W1'], x) # H x D * D X 1 = H x 1
h1 = act_1**2
act_2 = np.dot(model['W2'], h1)# 1 x H  *  H x 1 = 1 x 1 = Escalar (numbero)
out = act_2>0
out
```

### Matplotlib 

Liberia para graficar datos en 2, 3 dimensiones u otros tipos de visualizacion [matplotlib](http://matplotlib.org/users/beginner.html)


```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = [8, 5]

# Datos
x = np.arange(1,30,0.05)
y = np.sin(x)

# Graficar y 
plt.plot(x, y)
plt.show()
```

```python
# Configutacion
plt.plot(x, y, linewidth=3, linestyle="--",  color="magenta", label="sin(x)")
plt.xlabel(r"Angulo ($\theta$)")
plt.ylabel(r"Frecuencia($\psi$)")
plt.title("Hello World")
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm  # para esquema de colores
from mpl_toolkits.mplot3d import Axes3D # para poder graficar en 3D
import numpy as np
plt.rcParams['figure.figsize'] = [11, 6] # configuracion de tamaño

# datos en el plano 2D
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)

# datos en el plano 3D
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Graficar en 3D
fig = plt.figure() # obtener figura
ax = fig.gca(projection="3d") # obtener ejes
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm) # graficar datos x, y & z signando esquema de colores
plt.title('Grafico 3D', fontsize=20, fontweight='bold')
fig.colorbar(surf) # agregar indicador de esquema de colores
plt.show()

# Graficar como Isolineas (como los mapas indicando la altura de una motaña :D )
CS = plt.contour(X, Y, Z)
plt.clabel(CS, inline=1, fontsize=10) # configurar indicadores de cada isolinea
plt.title('Isolineas (2D)', fontsize=20, fontweight='bold')
plt.show()
```
